```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from tensorflow.keras.preprocessing import sequence
from sklearn.model_selection import train_test_split
from keras import layers
from keras.models import Model
```

```python
data = pd.read_csv("ner_dataset (1).csv", encoding="latin1")
data.head(50)
```

```python
data = data.fillna(method="ffill")
data.head(50)
```

```python
print("Unique words in corpus:", data['Word'].nunique())
print("Unique tags in corpus:", data['Tag'].nunique())
words=list(data['Word'].unique())
words.append("ENDPAD")
tags=list(data['Tag'].unique())
print("Unique tags are:", tags)
```

```python
num_words = len(words)
num_tags = len(tags)
num_words
```

```python
class SentenceGetter(object):
    def __init__(self, data):
        self.n_sent = 1
        self.data = data
        self.empty = False
        agg_func = lambda s: [(w, p, t) for w, p, t in zip(s["Word"].values.tolist(),
                                                           s["POS"].values.tolist(),
                                                           s["Tag"].values.tolist())]
        self.grouped = self.data.groupby("Sentence #").apply(agg_func)
        self.sentences = [s for s in self.grouped]
```

```python
def get_next(self):
        try:
            s = self.grouped["Sentence: {}".format(self.n_sent)]
            self.n_sent += 1
            return s
        except:
            return None
```

```python
getter = SentenceGetter(data)
sentences = getter.sentences
len(sentences)
sentences[0]
```

```python
word2idx = {w: i + 1 for i, w in enumerate(words)}
tag2idx = {t: i for i, t in enumerate(tags)}
word2idx
```

```python
print("balaji j ")
print("212221243001")
print("Length of each sentences")
```

```python
plt.hist([len(s) for s in sentences], bins=50)


```

```python
plt.show()
X1 = [[word2idx[w[0]] for w in s] for s in sentences]
```

```python
type(X1[0])
X1[0]
max_len = 55
```

```python
X = sequence.pad_sequences(maxlen=max_len,
                  sequences=X1, padding="post",
                  value=num_words-1)
X[0]



```

```python
y1 = [[tag2idx[w[2]] for w in s] for s in sentences]

y = sequence.pad_sequences(maxlen=max_len,
                  sequences=y1,
                  padding="post",
                  value=tag2idx["O"])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3, random_state=1)
X_train[0]

y_train[0]
```

```python
input_word = layers.Input(shape=(max_len,))
embedding_layer=layers.Embedding(input_dim=num_words,output_dim=55,input_length=max_len)(input_word)
dropout_layer=layers.SpatialDropout1D(0.2)(embedding_layer)
```

```python
bidirectional_lstm=layers.Bidirectional(
    layers.LSTM(units=100,return_sequences=True))(dropout_layer)
output=layers.TimeDistributed(
    layers.Dense(num_tags,activation="softmax"))(bidirectional_lstm)

```

```python
ent_name = Model(input_word, output)
ent_name.summary()
ent_name.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])
```

```python
history = ent_name.fit(
    x=X_train,
    y=y_train,
    validation_data=(X_test,y_test),
    batch_size=32,
    epochs=5,
)
```

```python
metrics = pd.DataFrame(ent_name.history.history)
metrics.head()
print("BALAJI J")
print("212221243001")
print("Accuracy VS validation accuracy")
metrics[['accuracy','val_accuracy']].plot()
print("BALAJI J")
print("212221243001")
print("Loss VS Validation loss")
metrics[['loss','val_loss']].plot()
```
